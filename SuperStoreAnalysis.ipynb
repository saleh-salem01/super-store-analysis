# What are the most selling sections?

```python
sell_section=data.groupby('Category')['Quantity'].sum()
```

```python
sell_section=pd.DataFrame(data.groupby('Category')['Quantity'].sum())
```

```python
sell_section
```

```python
import matplotlib.pyplot as plt
```

```python
x = sell_section.index
y = sell_section.Quantity
plt.bar(x,y,width = 0.6)
plt.xlabel("Category")
plt.ylabel("Quantity")
#plt.grid()
plt.show()
```

#### Conclusion: the office Supplies Category is the most selling section.

# the best city

```python
city_sell=data.groupby('City')['Quantity','Sales'].sum()
```

```python
city_sell.shape
```

```python
city_sell.head()
```

### The number of cities is too large and results are not clear when representing the data
### We can have the top 20 selling cities and they will be the best sellers

```python
maxCity=city_sell.nlargest(20, 'Quantity')
```

```python
Labels=maxCity.index
x= maxCity.Sales
#plt.title('the most City')
#plt.xlabel('City')
#plt.ylabel('Quantity')
plt.pie(x,labels=Labels,radius=2.3)
#plt.xticks(Labels,rotation=90)
plt.show()
```
